```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
Delhi_new_venues = getNearbyVenues(names=Delhi_new['place_name'],
                                   latitudes=Delhi_new['latitude'],
                                   longitudes=Delhi_new['longitude']
                                  )
```

```python
#Let us cluster based on neighborhoods
# one hot encoding
Delhi_new_onehot = pd.get_dummies(Delhi_new_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Delhi_new_onehot['Neighborhood'] = Delhi_new_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Delhi_new_onehot.columns[-1]] + list(Delhi_new_onehot.columns[:-1])
Delhi_new_onehot = Delhi_new_onehot[fixed_columns]

Delhi_new_grouped = Delhi_new_onehot.groupby('Neighborhood').mean().reset_index()
```

```python
Delhi_new_grouped.head(10)
```

```python
#Top common venues

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
import numpy as np
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
Delhi_new_venues_sorted = pd.DataFrame(columns=columns)
Delhi_new_venues_sorted['Neighborhood'] = Delhi_new_grouped['Neighborhood']

for ind in np.arange(Delhi_new_grouped.shape[0]):
    Delhi_new_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Delhi_new_grouped.iloc[ind, :], num_top_venues)

Delhi_new_venues_sorted.head()
```

## Apply Machine learning - clustering algorithm

Make the clusters of above neighbouhood so that client can choose the place of their choice eaisly

```python
# set number of clusters
kclusters = 4

Delhi_new_grouped_clustering = Delhi_new_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Delhi_new_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:24] 
```

```python
# add clustering labels
Delhi_new_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

Delhi_new_merged = Delhi_new

Delhi_new_merged = Delhi_new_merged.join(Delhi_new_venues_sorted.set_index('Neighborhood'), on='place_name')

Delhi_new_merged.head() # check the last c
```

```python
Delhi_new_merged['Cluster Labels'].replace(np.NaN, 0, inplace = True)
Delhi_new_merged['Cluster Labels']=Delhi_new_merged['Cluster Labels'].astype('int')

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Delhi_new_merged['latitude'], Delhi_new_merged['longitude'], Delhi_new_merged['community_name'], Delhi_new_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
